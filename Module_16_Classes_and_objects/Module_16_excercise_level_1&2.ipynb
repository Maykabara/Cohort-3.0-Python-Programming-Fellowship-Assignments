```python
from collections import Counter

class Statistics:
    def __init__(self, data):
        self.data = data

    def count(self):
        return len(self.data)

    def sum(self):
        return sum(self.data)

    def min(self):
        return min(self.data)

    def max(self):
        return max(self.data)

    def range(self):
        return self.max() - self.min()

    def mean(self):
        return self.sum() / self.count()

    def median(self):
        sorted_data = sorted(self.data)
        n = self.count()
        mid = n // 2
        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        else:
            return sorted_data[mid]

    def mode(self):
        freq = Counter(self.data)
        max_count = max(freq.values())
        modes = [key for key, value in freq.items() if value == max_count]
        return (modes[0], max_count)  # Returning the first mode and its count

    def variance(self):
        mean = self.mean()
        return sum((x - mean) ** 2 for x in self.data) / self.count()

    def std(self):
        return self.variance() ** 0.5

    def freq_dist(self):
        freq = Counter(self.data)
        total = self.count()
        return [(round((count / total) * 100, 1), key) for key, count in freq.items()]

    def describe(self):
        description = f"""
Count: {self.count()}
Sum: {self.sum()}
Min: {self.min()}
Max: {self.max()}
Range: {self.range()}
Mean: {round(self.mean(), 1)}
Median: {self.median()}
Mode: {self.mode()}
Variance: {round(self.variance(), 1)}
Standard Deviation: {round(self.std(), 1)}
Frequency Distribution: {self.freq_dist()}
"""
        return description



ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]

# Creating an instance of the Statistics class
data = Statistics(ages)

# Print results
print('Count:', data.count())  
print('Sum: ', data.sum())  
print('Min: ', data.min()) 
print('Max: ', data.max()) 
print('Range: ', data.range())  
print('Mean: ', round(data.mean(), 1))  
print('Median: ', data.median())  
print('Mode: ', data.mode())  
print('Standard Deviation: ', round(data.std(), 1))  
print('Variance: ', round(data.variance(), 1))  
print('Frequency Distribution: ', data.freq_dist())  

# Print the describe output
print(data.describe())
```

# Module 16 Assigment: Level 2
### 1. Create a class called PersonAccount. It has firstname, lastname, incomes, expenses properties and it has total_income, total_expense, account_info, add_income, add_expense and account_balance methods. Incomes is a set of incomes and its description. The same goes for expenses.

```python
class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = []  # List of dictionaries with income details
        self.expenses = []  # List of dictionaries with expense details

    # Calculate total income
    def total_income(self):
        return sum(income['amount'] for income in self.incomes)

    # Calculate total expenses
    def total_expense(self):
        return sum(expense['amount'] for expense in self.expenses)

    # Add income
    def add_income(self, description, amount):
        self.incomes.append({'description': description, 'amount': amount})

    # Add expense
    def add_expense(self, description, amount):
        self.expenses.append({'description': description, 'amount': amount})

    # Calculate account balance
    def account_balance(self):
        return self.total_income() - self.total_expense()

    # Display account information
    def account_info(self):
        info = f"""
        Account Holder: {self.firstname} {self.lastname}
        Total Income: ${self.total_income()}
        Total Expense: ${self.total_expense()}
        Account Balance: ${self.account_balance()}
        """
        return info



# Create a new account
person = PersonAccount("Muhammad", "Al-amin")

# Add incomes
person.add_income("Salary", 5000)
person.add_income("Freelance", 1500)

# Add expenses
person.add_expense("Rent", 1200)
person.add_expense("Groceries", 400)

# Display account information
print(person.account_info())
```
